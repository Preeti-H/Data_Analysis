# Bivariant Data Analysis

```python
import seaborn as sns
```

Getting the number of not assigned values of each column

```python
df.isna().sum()
```

Delete the not assigned values

```python
drop = df.dropna()
```

Check for not assigned values in columns

```python
drop.isna().sum()
```

```python
Plot the values
```

```python
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
plt.title('Fare vs Age(r = {0:0.2f})'.format(pearsonr(drop['Fare'],drop['Age'])[0]))
sns.scatterplot(x="Fare",y="Age",data = df,hue='Sex')
```

```python
plt.title('Fare vs Age(r = {0:0.2f})'.format(pearsonr(drop['Fare'],drop['Age'])[0]))
sns.scatterplot(x="Fare",y="Age",data = df,hue='Pclass')
```

```python
plt.title('Age vs Fare(r = {0:0.2f})'.format(pearsonr(drop['Age'],drop['Fare'])[0]))
sns.scatterplot(x="Age", y="Fare",hue="Pclass", style="Sex", data=df)
plt.show()
```

# Multivariant Data Analysis

```python
sns.relplot(x="Age",y="Fare",col="Pclass",hue="Sex",style="Sex",
            kind="scatter",data = df,size ="Sex",sizes=(80,200))
plt.show()
```

```python
sns.pairplot(df,hue ='Pclass',height=2)

plt.show()
```
